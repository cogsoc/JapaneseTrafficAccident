```python
accident2019 = pd.read_csv("Data/honhyo_2019.csv", encoding="Shift-JIS")
```

```python
# 緯度 means latitude but the type of value is degrees, minutes, seconds (DMS) 
# So, it is necessary to change from DMS to Decimal Degrees(DD)
accident2019['地点　緯度（北緯）']
```

```python
import numpy as np
df=pd.DataFrame()
df["latitude_deg"] = (accident2019['地点　緯度（北緯）']/10000000).agg(np.floor)
df["latitude_deg"] 
```

```python
df["latitude_min"] = ((accident2019['地点　緯度（北緯）']-df["latitude_deg"] *10000000)/100000).agg(np.floor)/60
df["latitude_min"]
```

```python
df["latitude_sec"] = (accident2019['地点　緯度（北緯）'] -(accident2019['地点　緯度（北緯）'] /100000).agg(np.floor)*100000)/1000/3600
df["latitude_sec"]
```

```python
# DMS format to DD format
df["longitude_deg"] = (accident2019['地点　経度（東経）']/10000000).agg(np.floor)
df["longitude_min"] = ((accident2019['地点　経度（東経）']-df["longitude_deg"] *10000000)/100000).agg(np.floor)/60
df["longitude_sec"] = (accident2019['地点　経度（東経）'] -(accident2019['地点　経度（東経）'] /100000).agg(np.floor)*100000)/1000/3600
```

```python
accident2019["latitude"]=df["latitude_deg"] +df["latitude_min"]+df["latitude_sec"]  
accident2019["longitude"]=df["longitude_deg"] +df["longitude_min"]+df["longitude_sec"]  
accident2019.head()
```

```python
accident2019chiba=accident2019[accident2019["都道府県コード"] == 44]
accident2019chiba.head()
```

```python
# Aim is to predict the place where the fatal accidents would happen.
```

```python
accident_points = accident2019.loc[:,['latitude','longitude']]
accident_points_list = accident_points.values.tolist()
```

```python
from folium.plugins import FastMarkerCluster
from folium.plugins import HeatMap

my_map = folium.Map(location=[35.269368,139.909740], zoom_start=10)

FastMarkerCluster(accident_points_list).add_to(my_map)
HeatMap(accident_points_list, radius=10, blur=5).add_to(my_map)

my_map.save("mapcluster_test")
```
